```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 100
h = 1/N
x = np.linspace(0, 1 , N, endpoint=True)
u0 = np.sin(np.pi * x)
r = 0.4
a = 1

k = r * h**2 / a

iters = int(1 / k) + 1
```

```python
us = [u0]

for iter in range(1, iters):
    new_u = np.copy(us[-1])
    for i in range(1, N-1):
            sub_u = (1-2*r) * us[-1][i] + r*(us[-1][i-1] + us[-1][i+1])
            new_u[i] = sub_u
    
    us.append(new_u)
```

```python
us = np.asarray(us)
time = np.linspace(0, iters * k)
iters_array = np.arange(iters)

plt.matshow(us.T, aspect=(iters/N)*0.1)
plt.title(r"$\alpha=0.1$")
plt.xticks(
    [iters_array[0], iters_array[int(iters/2)], iters_array[-1]], 
    [time[0], time[int(len(time)/2)], time[-1]]
)
```
