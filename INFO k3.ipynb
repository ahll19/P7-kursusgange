```python
from itertools import islice

def take(n, iterable):
    return dict(islice(iterable, n))

def split_symbols(symbols):
    if len(symbols) > 1:
        total = sum([symbols[key] for key in symbols])
        running = 0
        for i, key in enumerate(symbols):
            running += symbols[key]
            if running >= total/2:
                new_symbols = take(i+1, symbols.items())
                split_symbols(new_symbols)
    
    print(f"Done with {symbols}")

split_symbols(let_freq_sorted)
```
